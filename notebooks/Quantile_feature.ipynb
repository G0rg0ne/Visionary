```python
import pandas as pd
```

```python
df_train = pd.read_csv('../data/04_feature/tickets_train_data_v1.csv')
```

```python
od_airline_dbd_price_list = (
    df_train
    .groupby(
        ["origin", "destination", "airline", "days_before_departure"],
        as_index=False
    )
    .agg(
        price_list=("price", list),
        per_dbd_mean_price=("price", "mean"),
        )
    .sort_values("airline", ascending=False)
)
od_airline_dbd_price_list
```

```python
one_airline_example = od_airline_dbd_price_list[od_airline_dbd_price_list["airline"]=="Vueling"]
one_airline_example
```

```python
# Assuming your DataFrame is called df
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.line(one_airline_example['days_before_departure'], one_airline_example['per_dbd_mean_price'], alpha=0.6)
plt.xlabel('Days Before Departure')
plt.ylabel('Mean Price')
plt.title('Mean Price vs Days Before Departure')
plt.grid(True)
plt.show()
```

```python
df_perc = df_train.merge(
    od_airline_dbd_price_list[
        ["origin", "destination", "airline", "days_before_departure", "price_list", "per_dbd_mean_price"]
    ],
    on=["origin", "destination", "airline", "days_before_departure"],
    how="left"
)
df_perc
```

```python
import numpy as np

def dbd_percentile(price, price_list):
    """Compute percentile of price within the reference price list."""
    prices = np.asarray(price_list)
    return (prices <= price).mean()  # fraction of prices <= current price

df_perc["dbd_percentile"] = df_perc.apply(
    lambda r: dbd_percentile(r.price, r.price_list) if isinstance(r.price_list, list) else np.nan,
    axis=1
)
```
