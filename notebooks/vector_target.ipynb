```python
import pandas as pd
```

```python
df = pd.read_csv("../data/02_intermediate/sampled_dataset.csv")
df
```

```python
keys = ['origin', 'destination', 'airline', 'departure_time', 'departure_date', 'days_before_departure']
df = df.sort_values('price').drop_duplicates(subset=keys)
df['todays_price'] = df['price']
for i in range(1, 8):
    future_data = df[keys + ['price']].copy()
    future_data['days_before_departure'] = future_data['days_before_departure'] + i
    # Rename for the merge
    future_data = future_data.rename(columns={'price': f'price_{i}'})
    # Merge efficiently
    df = pd.merge(
        df,
        future_data,
        on=keys,
        how='left'
    )
# 6. Final Selection
# Select your features + targets
target_cols = ['todays_price'] + [f'price_{i}' for i in range(1, 8)]
final_df = df[['origin', 'destination', 'departure_date', 'days_before_departure'] + target_cols]
```

```python
target_cols = ['price_1', 'price_2', 'price_3', 'price_4', 'price_5', 'price_6', 'price_7']

# 2. Count non-null values row-by-row
# notna() turns values to True(1) or False(0). sum(axis=1) sums them horizontally.
valid_target_count = df[target_cols].notna().sum(axis=1)

# 3. Filter: Keep rows where the count is >= 5
df_clean = df[valid_target_count >= 5].copy()
```

```python
df_clean
```

```python
fill_cols = ['todays_price', 'price_1', 'price_2', 'price_3', 'price_4', 'price_5', 'price_6', 'price_7']

# 2. Apply Forward Fill row-by-row (axis=1)
# This checks each row. If it finds a NaN, it looks to the left and takes that value.
df_clean[fill_cols] = df_clean[fill_cols].ffill(axis=1)
```
